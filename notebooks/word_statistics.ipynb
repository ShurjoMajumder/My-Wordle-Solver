```python
import functools
import pandas as pd
import numpy as np
import seaborn as sns
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
word_list: list[str] = []
with Path("../data/word_list.txt").open() as f:
    word_list = f.read().split(' ')
if not word_list:
    print("Failed to load word list")
```

```python
char_tuples = [
    (word, *tuple(char for char in word))
    for word in word_list
]
char_tuples
```

```python
words_df = pd.DataFrame.from_records(char_tuples, columns=["word", "l1", "l2", "l3", "l4", "l5"])
words_df
```

```python
endsWithS = words_df[words_df["l5"]=='S'].count()["l5"]
everythingElse = words_df.count()["l1"]
```
